## load The Data 

```python
import pandas as pd 
import matplotlib.pyplot as plt 

data = pd.read_csv('Mall_Customers.csv')
print (data.head())
```

## Data Cleaning 

```python

print(data.describe())

print(data['Gender'].value_counts())

gender_counts = data['Gender'].value_counts()
print("Gender Distribution:\n", gender_counts)
```

## Basic Analysis

```python

print("Missing Values:\n", data.isnull().sum())


duplicates = data.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")


if duplicates > 0:
    data = data.drop_duplicates()
    print("Duplicates removed.")


print("Data Types:\n", data.dtypes)


if data['Gender'].dtype != 'category':
    data['Gender'] = data['Gender'].astype('category')
    print("Converted Gender column to 'category' type.")
```

## Visualization 

### Gender Distribution

```python

gender_counts.plot(kind='bar', color=['blue', 'pink'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

### Age Distribution 

```python

plt.hist(data['Age'], bins=10, color='green', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

### Annual VS Spending Score

```python

plt.scatter(data['Annual Income (k$)'], data['Spending Score (1-100)'], alpha=0.6, color='purple')
plt.title('Annual Income vs Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

### Age VS Spending Score

```python

plt.scatter(data['Age'], data['Spending Score (1-100)'], alpha=0.6, color='orange')
plt.title('Age vs Spending Score')
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.show()
```
